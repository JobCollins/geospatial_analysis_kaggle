```python
import folium
from folium import Marker, GeoJson
from folium.plugins import HeatMap

import pandas as pd
import geopandas as gpd
```

```python
toxic_releases = gpd.read_file("../geospatial_kaggle/toxic_release_pennsylvania/toxic_release_pennsylvania/toxic_release_pennsylvania.shp")
toxic_releases.head()
```

```python
stations = gpd.read_file("../geospatial_kaggle/PhillyHealth_Air_Monitoring_Stations/PhillyHealth_Air_Monitoring_Stations/PhillyHealth_Air_Monitoring_Stations.shp")
stations.head()
```

## Measuring Distance

To measure distances between points from two different GeoDataFrames, we first have to make sure that they use the same coordinate reference system (CRS).

```python
print(stations.crs)
print(toxic_releases.crs)
```

check the CRS to see which units it uses (meters, feet, or something else). In this case, EPSG 2272 has units of feet.

The code cell below calculates the distance (in feet) between a relatively recent release incident in `recent_release` and every station in the `stations` GeoDataFrame.

```python
# select one release incident in particular
recent_toxic_release = toxic_releases.iloc[360]

# measure distance from release to each station
distances = stations.geometry.distance(recent_toxic_release.geometry)
distances
```

```python
# mean distance to each station
print('mean distance to monitoring stations: {} feet'.format(distances.mean()))
```

```python
print('Closest monitoring station ({} feet):'.format(distances.min()))
print(stations.iloc[distances.idxmin()][["ADDRESS", "LATITUDE", "LONGITUDE"]])
```

## Creating a buffer

If we want to understand all points on a map that are some radius away from a point, the simplest way is to create a buffer.

The code cell below creates a GeoSeries two_mile_buffer containing 12 different Polygon objects. Each polygon is a buffer of 2 miles (or, 2*5280 feet) around a different air monitoring station.

```python
two_mile_buffer = stations.geometry.buffer(2*5280)
two_mile_buffer.head()
```

use folium.GeoJson() to plot each polygon on a map. Note that since folium requires coordinates in latitude and longitude, we have to convert the CRS to EPSG 4326 before plotting.

```python
# map with release incidents and monitoring stations
m = folium.Map(location=[39.9526,-75.1652], zoom_start=11)
HeatMap(data=toxic_releases[['LATITUDE', 'LONGITUDE']], radius=15).add_to(m)

for idx, row in stations.iterrows():
    Marker([row['LATITUDE'], row['LONGITUDE']]).add_to(m)

# plot each polygon on the map
GeoJson(two_mile_buffer.to_crs(epsg=4326)).add_to(m)

m
```

Now, to test if a toxic release occurred within 2 miles of any monitoring station, we could run 12 different tests for each polygon (to check individually if it contains the point).

But a more efficient way is to first collapse all of the polygons into a MultiPolygon object. We do this with the unary_union attribute.

```python
# turn groups of polygons into single multipolygon
my_union = two_mile_buffer.geometry.unary_union
print('Type:', type(my_union))

# show the mutlipolygon object
my_union
```

use the `contains()` method to check if the multipolygon contains a point. We'll use the release incident from earlier in the tutorial, which we know is roughly 3781 feet to the closest monitoring station.

```python
# closest station is less than two miles away
my_union.contains(toxic_releases.iloc[360].geometry)
```

But not all releases occured within two miles of an air monitoring station!

```python
# closest station is more than two miles away
my_union.contains(toxic_releases.iloc[358].geometry)
```
